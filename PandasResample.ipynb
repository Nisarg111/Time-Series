# New Section
Importing data from yfinance plotting it.
Data can be obtained from yfinance in 2 ways ticker and download. More details: https://chatgpt.com/share/3e741c9e-45cb-4b53-a0ae-7e54113fd06a
Resample of data: This is used to group the data and perform some mathematical operation on it : https://chatgpt.com/share/76b85973-ef37-4d11-84a8-005e6966d789

Additional info Data from yfinance has Date index so it can not be accessed Directly and if we want to access it we have first reset the columns then we can do it.
Date column is default X-axis for plotting.

```python



import yfinance as yf
import pandas as pd

df_Microsoft = yf.download('MSFT')
df_Microsoft.plot()
```

```python
#annual highest for 5 years
df_yearly_max = df_Microsoft['High'].resample('Y').max()
df_yearly_max.iloc[0:5]
df_yearly_min = df_Microsoft['Low'].resample('Y').min()
df_yearly_min.iloc[0:5]
```
